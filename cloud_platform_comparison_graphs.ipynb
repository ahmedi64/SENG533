```python
import matplotlib.pyplot as plt
import pandas as pd

# Create the dataset
data = {
    "Platform": ["AWS", "Google Colab", "Azure"],
    "Step Time (ms)": [61.29, 80.1, 36.5],
    "Epoch Time (s)": [33.87, 37.54, 17.13],
    "Total Time (s)": [1050, 1126.31, 513.81],
    "Test Loss": [6.539, 7.22, 7.92],
    "Test Accuracy (%)": [98.6, 98.78, 98.6],
    "CPU (Cores)": [2, 2, 2],
    "RAM (GB)": [4, 12.67, 14]
}

df = pd.DataFrame(data)

# Compute statistics
numeric_cols = df.columns[1:]
means = df[numeric_cols].mean()
medians = df[numeric_cols].median()
variances = df[numeric_cols].var()

print("Means:\n", means)
print("\nMedians:\n", medians)
print("\nVariances:\n", variances)

# Plotting
plt.figure(figsize=(14, 10))

# Subplot 1: Step Time
plt.subplot(2, 2, 1)
plt.bar(df["Platform"], df["Step Time (ms)"])
plt.axhline(means["Step Time (ms)"], color='red', linestyle='--', label='Mean')
plt.axhline(medians["Step Time (ms)"], color='green', linestyle=':', label='Median')
plt.title("Step Time (ms)")
plt.ylabel("Time (ms)")
plt.legend()

# Subplot 2: Epoch Time
plt.subplot(2, 2, 2)
plt.bar(df["Platform"], df["Epoch Time (s)"])
plt.axhline(means["Epoch Time (s)"], color='red', linestyle='--', label='Mean')
plt.axhline(medians["Epoch Time (s)"], color='green', linestyle=':', label='Median')
plt.title("Epoch Time (s)")
plt.ylabel("Time (s)")
plt.legend()

# Subplot 3: Total Time
plt.subplot(2, 2, 3)
plt.bar(df["Platform"], df["Total Time (s)"])
plt.axhline(means["Total Time (s)"], color='red', linestyle='--', label='Mean')
plt.axhline(medians["Total Time (s)"], color='green', linestyle=':', label='Median')
plt.title("Total Training Time (s)")
plt.ylabel("Time (s)")
plt.legend()

# Subplot 4: Accuracy and Loss
plt.subplot(2, 2, 4)
plt.plot(df["Platform"], df["Test Accuracy (%)"], marker='o', label='Accuracy')
plt.plot(df["Platform"], df["Test Loss"], marker='s', label='Loss')
plt.axhline(means["Test Accuracy (%)"], color='red', linestyle='--', label='Mean Accuracy')
plt.axhline(medians["Test Accuracy (%)"], color='green', linestyle=':', label='Median Accuracy')
plt.axhline(means["Test Loss"], color='orange', linestyle='--', label='Mean Loss')
plt.axhline(medians["Test Loss"], color='blue', linestyle=':', label='Median Loss')
plt.title("Test Accuracy and Loss")
plt.ylabel("Percentage / Loss")
plt.legend()

plt.tight_layout()
plt.show()
```
